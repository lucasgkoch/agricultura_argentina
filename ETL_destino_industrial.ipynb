Con los dataframes procesados y todos con la misma estructura, se procede a concatenarlos.

```python
df_cultivos = pd.concat([df_avena, df_centeno, df_girasol, df_maiz, df_mani, df_soja_aceitera, df_soja_balanceadora], ignore_index=True)
```

```python
df_cultivos
```

Se transforman las columnas en filas individuales, ya que de esa forma lo podremos usar en nuestro posterior analisis.

```python
df_cultivos = pd.melt(df_cultivos, id_vars=['año', 'producto',], var_name='mes', value_name='toneladas')
```

```python
df_cultivos
```

Como para algunos granos teniamos mas de una tabla (soja), vamos a tener registros con año y mes duplicados. procedemos a agrupar por año, mes y producto sumando toneladas.

```python
df_cultivos = df_cultivos.groupby(['año', 'producto', 'mes'], as_index=False)['toneladas'].sum()
```

Reemplazamos los nombres de los meses por sus respectivos numeros

```python
meses_a_numeros = {
    'enero': 1,
    'febrero': 2,
    'marzo': 3,
    'abril': 4,
    'mayo': 5,
    'junio': 6,
    'julio': 7,
    'agosto': 8,
    'septiembre': 9,
    'octubre': 10,
    'noviembre': 11,
    'diciembre': 12
}

df_cultivos['mes'] = df_cultivos['mes'].map(meses_a_numeros)
```

Combinamos las columas año y mes, dejando una unica columna llamada fecha con el formato AAAA/MM

```python
df_cultivos['fecha'] = df_cultivos['año'].astype(str) + '/' + df_cultivos['mes'].astype(str)

df_cultivos.drop(columns=['año', 'mes'], inplace=True)
```
